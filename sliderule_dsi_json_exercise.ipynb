## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python

# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON example, with file

+ demonstrates reading in a json file as a string and as a table
+ uses small sample file containing data about projects funded by the World Bank 

```python
# load json as string
data = json.load((open('data/world_bank_projects.json')))
```

```python
# load as Pandas dataframe
df = pd.read_json('data/world_bank_projects.json')
df 
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

# Find the 10 countries with most projects

```python
top_countries = df[['countryname' , 'project_name']]
top_countries.groupby('countryname').size().sort_values(ascending=False).head(10)
```

```python
top_countries.groupby('countryname').size().sort_values(ascending=False).head(10).plot()

plt.xticks(
    rotation=45,
    horizontalalignment='right',
    fontweight='light',
    fontsize='medium',
)
```

# Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
# Find the top 10 major project themes (using column 'mjtheme_namecode')

top_proj = json_normalize ( data , record_path=['mjtheme_namecode'])

top_proj.name.value_counts().sort_values(ascending=False).head(10)
```

```python
top_proj.name.value_counts().sort_values(ascending=False).head(10).plot()

plt.xticks(
    rotation=45,
    horizontalalignment='right',
    fontweight='light',
    fontsize='medium',
)
```

#   In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in

```python
q3 = json_normalize ( data , record_path=['mjtheme_namecode'])

q3 = q3.drop_duplicates()
```

```python
q3 = q3 [q3.name != '']
q3['code'] = q3.code.astype(int)
q3 = q3.sort_values('code')
q3 = q3.reset_index(drop=True)
q3
```
